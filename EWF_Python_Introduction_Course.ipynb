*Fill in the ... to define a function that prints whether the first given paramter is greater or equal than the second paramter - or not:*

```
def compare(x, y): # hier werden x und y schon von der Position in compare (x,y) definiert:
  if x >= y:
    print(f"{x} ist größer oder gleich {y}") # larger equal
  else:
    print(f"{x} ist kleiner als {y}") # smaller

print(compare(3,4))
```

```
def compare(x, y): # hier werden x und y schon von der Position in compare (x,y) definiert:
  if x >= y:
    print(f"{x} ist größer oder gleich {y}") # larger equal
  else:
    print(f"{x} ist kleiner als {y}") # smaller
a = 3
b = 5
print(compare(x=a, y=b))
```

So far "Einkommen" is treated as a nominal variable - how can we turn it into a categorical one?

```
import numpy as np
categories=["< 1000", "1000 - < 2000", "2000 - < 4000", "4000 - < 6000", "> 6000"]
bins = [-np.inf, 1000, 2000, 6000, np.inf]
```

Pandas can also help visualizing data.Try to create a normal bar plot for age.

```
df['einkommen_kategorie'] = pd.cut(df['einkommen'], bins=bins, labels=categories, right=False)
print(df.head())
```

Now, can we turn this into a histogram, to get an idea about the age distribution in our dataset?

```
# pd.cut erzeugt eine neue Spalte mit den Kategorien
df['einkommen_kat'] = pd.cut(df['einkommen'], bins=bins, labels=categories, right=False, include_lowest=True)

print("\nAnzahl pro Einkommenskategorie:")
print(df['einkommen_kat'].value_counts().sort_index())
```
